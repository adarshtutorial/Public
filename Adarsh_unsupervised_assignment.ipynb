```python
#1. choose k values or number of clusters k=3
#2.select random centroids

X=data
K=3
centroids=(X.sample(n=K))
plt.scatter(X['Alcohol'],X['Malic_Acid'],c='black')
plt.scatter(centroids['Alcohol'],centroids['Malic_Acid'],c='red')
plt.xlabel('Alcohol in X axis')
plt.ylabel('Malic_Acid in Y axis')
plt.show()
```

```python
from sklearn.cluster import KMeans
K=3
kmeans=KMeans(n_clusters=K,random_state=0)
X["Cluster"]=kmeans.fit_predict(X[["Alcohol","Malic_Acid"]])
```

```python
#Get Cluster methods
centroids=kmeans.cluster_centers_

sns.scatterplot(data=X,x="Alcohol",y="Malic_Acid",hue="Cluster",palette="Set1")
plt.scatter(centroids[:, 0],centroids[:, 1],c='red',marker='x',s=100)
plt.xlabel('Alcohol')
plt.ylabel('Malic_Acid')
plt.show()
```

```python
inertia=kmeans.inertia_
print("Inertia",inertia)
```

```python
K_range=range(1,11)
inertia_values=[]

for K in K_range:
  kmeans=KMeans(n_clusters=K,random_state=0)
  kmeans.fit(X[["Alcohol","Malic_Acid"]])
  inertia=kmeans.inertia_
  inertia_values.append(inertia)

plt.plot(K_range,inertia_values,marker='o')
plt.title("Elbow method for Optimal K ")
plt.xlabel('No of Cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
#we will load the dataset  Then, we will select three columns 
from sklearn.cluster import DBSCAN
X_train = data[['Alcohol', 'Malic_Acid', 'Color_Intensity']]
```

```python
#eps: it is the radius of specific neighborhoods. If the distance between two points is less than or equal to esp, it will be considered its neighbors.
#minPts: minimum number of data points in a given neighborhood to form the clusters.  
clustering = DBSCAN(eps=0.5, min_samples=2).fit(X_train)
DBSCAN_dataset = X_train.copy()
DBSCAN_dataset.loc[:,'Cluster'] = clustering.labels_ 
```
