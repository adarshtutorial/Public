```python
target=dfnew1['label']
#df1.drop('label',axis=1,inplace=True)

#splitting and training
from sklearn.model_selection import train_test_split
#X=df1
X=dfnew1['text']
y=target
```

```python
#Scikit-learn algorithm (sklearn. model_selection. train_test_split), is recommended to used the parameter 
#( random_state=42) to produce the same results across a different run
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
#split the dataframe
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

# Assigning variable to the vectorizer

train_vectors = vectorizer.fit_transform(X_train)

test_vectors = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.model_selection import train_test_split, GridSearchCV

#random forest using three algorithms ..one such is entropy
#GINI
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier(n_estimators=400,criterion='entropy')



# Fit the random search object to the data


rf_model=rf.fit(train_vectors,y_train)

rf_predict=rf_model.predict(test_vectors)
accuracy = accuracy_score(y_test, rf_predict)*100
print('Accuracy of RandomForestClassifier model is equal ' + str(round(accuracy, 2)) + ' %.')
```

```python
#n_estimators=400- accuracy : 91.16%   criterion='entropy'
#n_estimators=400- accuracy : 91.4%   criterion='log_loss'


##n_estimators=350- accuracy :90.6 %

##n_estimators=450- accuracy : 91.4%
```

```python
#Naive Bayes classifier
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB(
    alpha=1.0
)
multinom_model=model.fit(train_vectors,y_train)

multinom_predict=multinom_model.predict(test_vectors)
accuracy = accuracy_score(y_test, multinom_predict)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

```python
#create and train a support vector classfier
from sklearn.svm import SVC
#create and train a support vector classfier
svc_classifier=SVC()
svc_model=svc_classifier.fit(train_vectors,y_train)

svc_predictions = svc_model.predict(test_vectors)

# Evaluate the accuracy of the classifiers
svc_accuracy = accuracy_score(y_test, svc_predictions)*100

#accuracy of SVC classifier is below
#print("Support Vector Classifier Accuracy:", svc_accuracy)

print('accuracy of Support Vector Classifier Accuracy  model is equal to '+ str(round(svc_accuracy,2))+'%.')
```
