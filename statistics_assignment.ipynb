```python
#Testing of Hypothesis
data1=pd.read_csv(r"D:\ad\tutorials\Data-analys\class-11-statistics\assignment_statistics\Sales_add.csv")
data1.head()

```

```python
data1.info()
```

```python
data1.describe()
```

```python
from scipy.stats import ttest_ind

salesbeforedigital=data1['Sales_before_digital_add(in $)']
salesafterdigital=data1['Sales_After_digital_add(in $)']

t_stat1,pval1=ttest_ind(salesbeforedigital,salesafterdigital)
```

```python
if(pval1<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
#Here Null hypothesis is no increase in sales after stepping into digital marketing
#HA- increase in sales after stepping into digital marketing
#Here clearly sales got increased after stepping into digital marketing .So Null hypothesis rejected
```

```python
#H0-null hypothesis: there is no relation between the features Region and Manager
#Alternate hypothesis: there is significant relation between these two
#Performing  chi-square contingency test
from scipy.stats import chi2_contingency
contingency=pd.crosstab(data1['Region'],data1['Manager'])
contingency
```

```python
chi2_stat2,p_value2,dof,exp=chi2_contingency(contingency)
chi2_stat2,p_value2,dof,exp
```

```python
if(p_value2<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
    
```

```python
# So there is no relationship between the features region and Manager as null hypothesis is accepted
```
